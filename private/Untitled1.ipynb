```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arrow
plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
x = y = np.linspace( -1,1,11 )
for i in range(x.size):
    for j in range(y.size):
        rmag = np.hypot(x[i],y[j])
        efield = np.array( [ 1/rmag**3*x[i],1/rmag**3*y[j] ] )
        efield = efield / np.linalg.norm(efield) / 5
        arrow = Arrow( x[i],y[j],efield[0],efield[1],width=.1 )
        plt.gca().add_artist( arrow )

plt.xlim(-1,1)
plt.ylim(-1,1)
plt.gca().set_aspect('equal')
```

```python
rstart = [1,1]
r = list( rstart )
x = [r[0]]
y = [r[1]]
while (np.abs(r[0]-0.5) > 0.1) and (np.abs(r[1]+1) > 0.1):
    dx = np.random.uniform(-.5,.5)
    dy = np.random.uniform(-.5,.5)
    r[0] += dx
    r[1] += dy
    x.append( r[0] )
    y.append( r[1] )
```

```python
plt.plot(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arrow
#plt.style.use("/Users/tjwilli/jupyter.mplstyle")
```

```python
plt.figure(figsize=(16,10))
ncharges = 10
#theta = np.linspace(0,np.pi,ncharges)
#theta2 = theta[1:-1] - np.pi
#theta = np.append(theta,theta2)
theta = np.arange(0,2*np.pi,np.pi/10)

x = np.cos(theta)
y = np.sin(theta)
plt.scatter(x,y,zorder=2)
circ = plt.Circle(xy=(0,0),radius=1,fill='')
plt.gca().add_artist(circ)
plt.gca().set_aspect('equal')
eobs = np.array( [0,0] )
etot = 0
NORM = 3
for i in range(theta.size):
    ix = x[i]
    iy = y[i]
    rsrc = np.array( [ix,iy] )
    robs = eobs
    r = robs - rsrc
    rmag = np.dot(r,r)**.5
    efield = 1 / rmag**3 * r
    efield = efield / NORM
    #print(efield,np.linalg.norm(efield),rmag)
    etot+=efield
    #print(etot)
    arrow = Arrow(robs[0],robs[1],efield[0],efield[1],width=.075)
    #arrow = Arrow(rsrc[0],rsrc[1],r[0],r[1],width=.1,linestyle='--')
    plt.gca().add_patch(arrow)    
#plt.axvline(eobs[0],c='k',ls='--')
#plt.axhline(eobs[1],c='k',ls='--')
plt.title('E at center of spherical shell')
```
